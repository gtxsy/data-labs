```python
def even(x):
    return(x % 2 == 0)

# Calculate odds of rolling and even-sided dice
sample_space = [1, 2, 3, 4, 5, 6]
die_sides = len(sample_space)
even_roll = len([x for x in sample_space if even(x)])
even_probability = even_roll / die_sides
print(even_probability)
```

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt

%matplotlib inline

# Standard Uniform Distribution in Python (CONTINUOUS)
fig, ax = plt.subplots(1, 1)
x = np.linspace(0,1, 100)
ax.plot(x, uniform.pdf(x))
```

```python
import math
from scipy.stats import norm

# Normal Distribution in Python (CONTINUOUS)
normal = norm(0, math.sqrt(9)) 
print(normal.pdf(4))
print(normal.cdf(2))       
print(normal.rvs()) 
fig, ax = plt.subplots(1, 1)
x = np.linspace(-3,3,1000)
y = norm.pdf(x)
ax.plot(x,y)
plt.show()
```

```python
from scipy.stats import expon

# Exponential Distribution in Python (CONTINUOUS)
exp = expon(1) 
print(exp.pdf(4))
print(exp.cdf(2))       
print(exp.rvs()) 
x = np.linspace(0,10,100)
y = expon.pdf(x)
fig, ax = plt.subplots(1, 1)
ax.plot(x,y)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Bernoulli Distribution Example (DISCRETE)
p = 0.8
X = bernoulli.rvs(p, size=100)
plt.hist(X, align='mid')
```

```python
from scipy import stats
from scipy.stats import binom

# Binomial Distribution (DISCRETE)
n=10
p=0.4
binomial = binom(n,p)

print(binomial.pmf(3))
print(binomial.cdf(3))
print(binomial.mean())
print(binomial.var())
print(binomial.std())
print(binomial.rvs(10))
```

```python
# Probability Mass Function (PMF) for Binomial Distribution
x = np.arange(0,10)
fig, ax = plt.subplots(1, 1)
ax.plot(x, binom.pmf(x, n, p), 'bo')
ax.vlines(x, 0, binom.pmf(x, n, p), colors='b', lw=5, alpha=0.5)
```

```python
from scipy import stats
from scipy.stats import poisson

# Poisson Distribution (DISCRETE)
param = 2
po = stats.poisson(param)
print(po.pmf(param))     
print(po.rvs(10))
```

```python
x = np.arange(0,6)
fig, ax = plt.subplots(1, 1)
ax.plot(x, poisson.pmf(x, param), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, poisson.pmf(x, param), colors='b', lw=5, alpha=0.5)
```
