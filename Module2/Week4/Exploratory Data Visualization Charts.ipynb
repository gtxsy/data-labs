```python
import pandas as pd

# How to create a box-plot with pandas
data = pd.read_csv('./data/vehicles.csv')
data.boxplot(column="Combined MPG") # this needs to run twice on my system to display the boxplot
```

```python
import seaborn as sns

# How to create a box-plot with seaborn
ax = sns.boxplot(x="Fuel Type", y="Highway MPG", data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

```python
cyl = [6.0,4.0]
data.Cylinders.isin(cyl)
data1 = data[data.Cylinders.isin(cyl)]
data1.shape
```

```python
data1.head()
```

```python
# Create a box-plot from a subset of data [data1]
ax = sns.boxplot(x="Fuel Type", y="Highway MPG", hue="Cylinders",data=data1, palette="Set3")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```

```python
# Create a violin-plot with seaborn
sns.violinplot("Combined MPG", data=data)
```

```python
# Create a histogram with pandas
data['Fuel Barrels/Year'].hist()
```

```python
# Create a histogram with pandas, with 50 bins
data['Fuel Barrels/Year'].hist(bins=50)
```

```python
# Create a histogram with seaborn
sns.distplot(data['Fuel Barrels/Year'])
```

```python
# Create a histogram with seaborn, with 20 bins
sns.distplot(data['Fuel Barrels/Year'], bins=20)
```

```python
# Create a bar chart with pandas
data['Drivetrain'].value_counts().plot.bar()
```

```python
# Create a bar chart with seaborn
ax = sns.barplot(x="Fuel Type", y="Highway MPG", hue="Cylinders", data=data1)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
```
