```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
var1 = np.arange(0,10,1) 
var1
```

```python
var2 = np.linspace(0,5,10)
var2
```

```python
# Draw a linear line plot (maps x1 to x1; x=y) with matplotlib
x1 = np.arange(0,10,1)
plt.plot(var1, var1, label='linear')
```

```python
# Draw a quadradic line plot with matplotlib
plt.plot(var1, var2**2, label='quadratic')
```

```python
plt.plot(var1, var1, label='linear')
plt.plot(var1, var2**2, label='quadratic')

# How to label the plot
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Linear vs Non Linear Plots")
plt.legend()

plt.show()
```

```python
# How to make sub-plots with matplotlib
fig, [ax1, ax2] = plt.subplots(1,2)

# Create plot 1, ax1
ax1.plot(var1,var2)
ax1.set_title('LINEAR PLOT')   
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')   

# Create plot 2, ax2
x = np.arange(0,10,0.1)
y = np.cos(x)
ax2.plot(x,y)
ax2.set_title('COSINE PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
plt.show()
```

```python
# How to make scatter plot with matplotlib
x = np.random.randn(10)
y = np.random.randn(10)
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
# How to make a pie chart with matplotlib
# WARNING: pie charts are bad
# https://www.businessinsider.com/pie-charts-are-the-worst-2013-6
labels = ['Vegetarians', 'Non-Vegetarians', 'Vegans']
sizes = [40, 45, 15]

explode = (0, 0.1, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Set Seaborn "aesthetic": https://seaborn.pydata.org/generated/seaborn.set.html
sns.set()
```

```python
# Draw a basic [quadratic] lineplot with seaborn
data = pd.DataFrame()
data['X'] = np.arange(0,10,1)
data['Y'] = data['X']**2
sns.lineplot(data.X, data.Y)
```

```python
# Draw a scatter plot with seaborn lmplot()
sns.lmplot("X", "Y", data=data, fit_reg=False)
```

```python
# Draw plot for a DataFrame with seaborn scatterplot()
data = pd.read_csv('./data/vehicles.csv') 
sns.scatterplot(x="CO2 Emission Grams/Mile", y="Highway MPG", hue="Drivetrain", data=data)
```

```python
# Combine scatterplot and lineplot with seaborn relplot()
sns.relplot(x='CO2 Emission Grams/Mile', y='Highway MPG', hue="Fuel Type", size="Fuel Type", data=data)
```

```python
# Draw relationship between categorical and numeric data with seaborn catplot()
sns.catplot(x="Highway MPG", y="Drivetrain", hue="Fuel Type", data=data)
```
