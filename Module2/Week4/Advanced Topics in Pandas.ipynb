```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
cols = ['mpg','cylinders','displacement','horsepower','weight','acceleration',
        'model_year','origin','car_name']
auto = pd.read_csv(url, sep='\\s+', names=cols)
auto.head()
```

```python
auto.index
```

```python
from pandas import RangeIndex

# Get all EVEN rows with RangeIndex
auto.index = RangeIndex(start=0, stop=398 * 2, step=2)
auto.head()
```

```python
auto.index = auto.car_name # make this column the new index
auto.head()
```

```python
auto2 = auto.copy()
auto2.set_index('car_name', inplace=True) # make this column the new index, and remove old column
auto2.head()
```

```python
# Count all name-model combinations with value_counts
(auto.car_name + auto.model_year.map(str)).value_counts()
```

```python
# Drop rows containing duplicates entries in these 2 columns
auto.drop_duplicates(subset=['car_name', 'model_year'], inplace=True)
auto.head()
```

```python
auto.set_index(['car_name', 'model_year'], inplace=True)
auto.head()
```

```python
new_index = [('fiat punto', 71)] + list(auto.index)
auto_reindexed = auto.reindex(new_index)
auto_reindexed.head()
```
