```python
import requests

url = 'https://www.reuters.com/article/us-shazam-m-a-apple-eu/eu-clears-apples-purchase-of-shazam-idUSKCN1LM1TZ'
html = requests.get(url).content

# Show first 600 chars from given webpage
html[0:600]
```

```python
from bs4 import BeautifulSoup

# Use BeautifulSoup library to read the raw HTML
soup = BeautifulSoup(html, "lxml")
soup
```

```python
# Get all text contained in header 'hn' tags
tags = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'p']
text = [element.text for element in soup.find_all(tags)]
text
```

```python
print('\n'.join(text))
```

```python
# Extract the table containing each country's life expectancy, from given wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_European_countries_by_life_expectancy'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
table = soup.find_all('table',{'class':'sortable wikitable'})[0]
table
```

```python
# How to format table from Wikipedia into lists for a DataFrame
rows = table.find_all('tr')
rows = [row.text.strip().split("\n") for row in rows]
rows[0:5]
```

```python
import pandas as pd

for row in rows:
    try:
        row.remove('')
        row.remove('')
    except:
        continue

colnames = rows[0]
data = rows[1:]

df = pd.DataFrame(data, columns=colnames)
df.head()
```

```python
# How to handle HTML status responses
import requests
r = requests.get(url) # `url` has been defined before
if r.status_code < 300:
    print('request was successful')
elif r.status_code >= 400 and r.status_code < 500:
    print('request failed because the resource either does not exist or is forbidden')
else:
    print('request failed because the response server encountered an error')
```

```python
# How to handle request errors
try:
    r = requests.get(url, timeout=10)
    print(r)
except requests.exceptions.Timeout:
    print('timeout error... do something')
except requests.exceptions.TooManyRedirects:
    print('redirect error... do something')
except requests.exceptions.SSLError:
    print('ssl error... do something')
except requests.exceptions.RequestException as e:
    print('other unknown errors... do something')
```

```python
# How to make an asynchronus request
import asyncio, requests
import nest_asyncio

nest_asyncio.apply()

urls = [
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/breakfast.jpg',
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/forbidden',
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/the-html5-breakfast-site.html'
]

async def main():
    loop = asyncio.get_event_loop()
    futures = [loop.run_in_executor(None, requests.get, url) for url in urls]
    for response in await asyncio.gather(*futures):
        print(response.status_code)

loop = asyncio.get_event_loop()
loop.run_until_complete(main())
```

```python
# How to deal with Throttling and Rate Limits
import requests, time

urls = [
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/breakfast.jpg',
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/forbidden',
    'https://s3-eu-west-1.amazonaws.com/ih-materials/uploads/data-static/documents/the-html5-breakfast-site.html'
]

for url in urls:
    response = requests.get(url)
    print(response.status_code)
    time.sleep(1)
```
