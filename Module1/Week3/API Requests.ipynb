```python
# Nested response from "GitHub Events" API
response = requests.get('https://api.github.com/events')

data = pd.DataFrame(response.json())
data.head(10)
```

```python
# Iterate through the columns in the API response and:
# 1. Turn the nested dictionaries into a data frame with a column for each key
# 2. Assign column names to each column in this new data frame
# 3. Add these new columns to the original data frame
# 4. Drop the column with the nested dictionaries
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
nested_columns = ['actor', 'org', 'payload', 'repo']

flat = flatten(data, nested_columns)
flat.head(10)
```
