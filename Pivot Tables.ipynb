```python
import pandas as pd
import numpy as np

vehicles = pd.read_csv('./data/vehicles.csv')
```

```python
# Get a pivot table where the index is vehicle class
vehicles.pivot_table(index=["Vehicle Class"]).head()
```

```python
# Group DF based on Vehicle Class and show values for each unique value in 'Cylinders'
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"]).head()
```

```python
# Map 'Combined MPG' to each 'Vehicle Class' by aggrgating the values
vehicles.pivot_table(index=["Vehicle Class"], values=["Combined MPG"], aggfunc=np.sum).head()
```

```python
# Map 2 columns to the 'Vehicle Class' index
vehicles.pivot_table(index=["Vehicle Class"], values=["Combined MPG", "CO2 Emission Grams/Mile"]).head()
```

```python
# Replace empty (NaN) values with 0 using fill_value=
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"], values=["Combined MPG"], fill_value=0).head()
```

```python
# Compute the correlation (Pearson correlation) between city MPG and highway MPG
vehicles['City MPG'].corr(vehicles['Highway MPG'])
```

```python
# Compute correlation (Pearson correlation) matrix for each pair of columns in the DataFrame
vehicles.corr()
```

```python
# Compute Spearman correlation
vehicles.corr(method='spearman')
```

```python
# Compute Kendall's Tau correlation coefficient
vehicles['City MPG'].corr(vehicles['Highway MPG'], method="kendall")
```
