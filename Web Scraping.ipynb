```python
import requests

url = 'https://www.reuters.com/article/us-shazam-m-a-apple-eu/eu-clears-apples-purchase-of-shazam-idUSKCN1LM1TZ'
html = requests.get(url).content

# Show first 600 chars from given webpage
html[0:600]
```

```python
from bs4 import BeautifulSoup

# Use BeautifulSoup library to read the raw HTML
soup = BeautifulSoup(html, "lxml")
soup
```

```python
# Get all text contained in header 'hn' tags
tags = ['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'h7', 'p']
text = [element.text for element in soup.find_all(tags)]
text
```

```python
print('\n'.join(text))
```

```python
# Extract the table containing each country's life expectancy, from given wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_European_countries_by_life_expectancy'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
table = soup.find_all('table',{'class':'sortable wikitable'})[0]
table
```

```python
# How to format table from Wikipedia into lists for a DataFrame
rows = table.find_all('tr')
rows = [row.text.strip().split("\n") for row in rows]
rows[0:5]
```

```python
import pandas as pd

for row in rows:
    try:
        row.remove('')
        row.remove('')
    except:
        continue

colnames = rows[0]
data = rows[1:]

df = pd.DataFrame(data, columns=colnames)
df.head()
```

```python
import re

# Example: How to find tags with regex
url = 'http://uscode.house.gov/download/download.shtml'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")
pattern = 'us/usc/\w*'
table = soup.find_all('div',{'class':'usctitle', 'id': re.compile(pattern)})
table
```
